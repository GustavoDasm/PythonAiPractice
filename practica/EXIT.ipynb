```python
#Problema
#Prediccion que si el grupo de kpop tendra exito o no tiempo de ensayo, presupuesto que se le da al grupo
#si es agradable para el publico(cantidad de estrellitas recien al inicio de su primer comeback)(vistas en youtube) gnancias(al mes)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
num_grupos = 100

horas_semanales_ensayo = np.random.randint(1, 11, num_grupos)
presupuesto_debut = np.random.randint(10000, 100001, num_grupos)
publicidad = np.random.randint(1, 51, num_grupos)
vistas_youtube = np.random.randint(1000, 50001, num_grupos)
ganancias_mes = presupuesto_debut * np.random.uniform(0.1, 0.7, num_grupos)

dataset = pd.DataFrame({
    'Hour/week de ensayo': horas_semanales_ensayo,
    'Presupuesto Debut': presupuesto_debut,
    'Publicidad': publicidad,
    'Vistas Youtube': vistas_youtube,
    'Ganancias del Mes': ganancias_mes
})

print(dataset.head(50))
```

```python
dataset.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Visualizar las relaciones entre las variables
plt.figure(figsize=(10, 6))
plt.scatter(dataset['Hour/week de ensayo'], dataset['Ganancias del Mes'])

plt.xlabel('Horas de Ensayo')
plt.ylabel('Ganancias Mensuales')
plt.title('Relación entre Horas de Ensayo y Ganancias Mensuales')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = dataset[['Horas_Semanales_Ensayo']]
y = dataset['Ganancias_Mes']
model.fit(X, y)

# Predecir los valores de 'Ganancias_Mes' utilizando el modelo
y_pred = model.predict(X)

# Visualizar los datos y la línea de regresión
plt.figure(figsize=(10, 6))
plt.scatter(dataset['Horas_Semanales_Ensayo'], dataset['Ganancias_Mes'])
plt.plot(X, y_pred, color='red') # Agregar la línea de regresión
plt.xlabel('Horas de Ensayo')
plt.ylabel('Ganancias Mensuales')
plt.title('Relación entre Horas de Ensayo y Ganancias Mensuales con Regresión Lineal')
plt.show()
```

```python
# Paso 5  Realizar predicciones con nuevos datos
try:
    grupo_id = int(input("Ingrese el id del grupo que quieres consultar: "))
    horas_ensayo = dataset[dataset['Horas_Semanales_Ensayo'] == grupo_id]['Horas_Semanales_Ensayo'].values[0]
    exito_pred = model.predict([[horas_ensayo]])
    success_percentage = (exito_pred[0] / 50000) * 100
    if success_percentage > 50:
        exito = "Yes"
    else:
        exito = "No"
    print(f"Predicción para el grupo {grupo_id}: {exito}")
except ValueError:
    print("Por favor, ingrese el id del grupo.")
except IndexError:
    print("No se encontró un grupo con ese ID.")
```

```python
# Paso 5  Realizar predicciones con nuevos datos
exitos = []
for grupo_id in dataset['Horas_Semanales_Ensayo'].unique():
    horas_ensayo = dataset[dataset['Horas_Semanales_Ensayo'] == grupo_id]['Horas_Semanales_Ensayo'].values[0]
    exito_pred = model.predict([[horas_ensayo]])
    success_percentage = (exito_pred[0] / 50000) * 100
    if success_percentage > 50:
        grupo_info = dataset[dataset['Horas_Semanales_Ensayo'] == grupo_id]
        exitos.append((grupo_id, grupo_info, success_percentage))

if exitos:
    print("Grupos con éxito:")
    for grupo_id, grupo_info, success_percentage in exitos:
        print(f"Grupo ID: {grupo_id}")
        print(grupo_info)
        print(f"Porcentaje de éxito: {success_percentage:.2f}%")
        print("-" * 20)
else:
    print("No se encontraron grupos con éxito.")
#try:
#    grupo_id = int(input("Ingrese el id del grupo que quieres consultar: "))
#    horas_ensayo = dataset[dataset['Horas_Semanales_Ensayo'] == grupo_id]['Horas_Semanales_Ensayo'].values[0]
#    exito_pred = model.predict([[horas_ensayo]])
#    success_percentage = (exito_pred[0] / 50000) * 100
#    if success_percentage > 50:
#        exito = "Yes"
#    else:
#        exito = "No"
#    print(f"Predicción para el grupo {grupo_id}: {exito}")
#except ValueError:
#    print("Por favor, ingrese el id del grupo.")
#except IndexError:
#    print("No se encontró un grupo con ese ID.")
```
