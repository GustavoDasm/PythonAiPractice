```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as graf
```

```python
#Suma de vectores
vector1 = np.random.randint(1, 5, size=9)
vector2 = np.random.randint(1, 5, size=9)
sumaVectores = vector1 + vector2

print("Vector 1\n", vector1)
print("Vector 2\n", vector2)
print("Suma\n", sumaVectores)

graf.figure()
grafica = graf.gca()

grafica.quiver(0, 0, vector1[0], vector1[1], angles="xy", scale_units="xy", scale=1, color="b", label=f"Vector 1: {vector1}")
grafica.quiver(0, 0, vector2[0], vector2[1], angles="xy", scale_units="xy", scale=1, color="g", label=f"Vector 2: {vector2}")
grafica.quiver(0, 0, sumaVectores[0], sumaVectores[1], angles="xy", scale_units="xy", scale=1, color="r", label=f"Suma de Vectores: {sumaVectores}")
grafica.set_xlim([0,10])
grafica.set_ylim([0,10])

graf.xlabel("Eje x")
graf.ylabel("Eje y")
graf.grid()
graf.legend()
graf.show()
```

```python
#Producto de Matrices
matriz1 = np.random.randint(1, 4, size=(4, 4))
matriz2 = np.random.randint(1, 4, size=(4, 4))

print("Matriz 1\n", matriz1)
print("Matriz 2\n", matriz2)

producto_Matrices = np.dot(matriz1, matriz2)
print("Producto de matrices\n", producto_Matrices)
```

```python
#Producto Escalar 
matriz = np.random.randint(1, 4, size=(4, 4))
escalar = 5

print("Matriz\n", matriz)
print("Escalar\n", escalar)

producto_Escalar = matriz * escalar
print("Producto de matrices\n", producto_Escalar)
```

```python
#Sistema de ecuacion lineal
variables_nose = np.random.randint(1, 6, size=(2, 2))
resultados = np.random.randint(-5, 5, size=2)

print("Matriz\n", variables_nose)
print("Vector\n", resultados)

solucion = np.linalg.solve(variables_nose, resultados)
print(f"Solucion\nSolucion 1 = {solucion[0]}\nSolucion 2 = {solucion[1]}" )
```
